```python
import csv

def csv_to_json(csv_file_path):
    data = []
    with open(csv_file_path, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)  # Reads each row as a dictionary
        for row in csv_reader:
            data.append(row)  # Add each row dictionary to the list
    return data
```

```python
coffee_imports = csv_to_json("../coffee-data/public/data/Coffee_import.csv")

for i in coffee_imports:
  i["Country"] = i["Country"].strip()

coffee_imports
```

```python
def find_non_matching_countries(countries_coffee, countries_geojson):
    non_matching_countries = []
    for country in countries_coffee:
        if country not in countries_geojson:
            non_matching_countries.append(country)
    return non_matching_countries
```

```python
non_matching_countries_imports = find_non_matching_countries([i["Country"] for i in coffee_imports], countries_geojson)

non_matching_countries_imports
```

```python
countries_geojson

new_country_names_imports = {
  "Czechia": "Czech Republic",
  "United States of America": "United States",
}
```

```python
def rename_and_organize_countries(data, country_names):
    # Rename countries in data based on the country_names mapping
    for country in data:
        if country["Country"] in country_names:
            country["Country"] = country_names[country["Country"]]

    # Organize data into a dictionary with country names as keys
    renamed_countries = {item["Country"]: item for item in data}
    return renamed_countries
```

```python
renamed_countries_coffee_imports = rename_and_organize_countries(coffee_imports, new_country_names_imports)
renamed_countries_coffee_imports
```

```python
import json

def dict_to_geojson(dictionary, filename):
    """
    Converts a dictionary to a GeoJSON file.

    Parameters:
    dictionary (dict): The dictionary to convert. Must follow GeoJSON structure.
    filename (str): The filename for the GeoJSON output (e.g., 'output.geojson').

    Returns:
    None
    """
    # Ensure the dictionary is structured as a GeoJSON FeatureCollection
    if dictionary.get("type") != "FeatureCollection":
        raise ValueError("Dictionary must be a valid GeoJSON FeatureCollection.")

    # Write to GeoJSON file
    with open(filename, 'w') as geojson_file:
        json.dump(dictionary, geojson_file, indent=4)
    print(f"GeoJSON file saved as {filename}")

new_data_to_geojson_format = {
  "type": "FeatureCollection",
  "features": new_data
}

dict_to_geojson(new_data_to_geojson_format, '../coffee-data/public/data/coffee_data.geojson')
```
